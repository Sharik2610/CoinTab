#**Importing Libraries**

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

#**Loading DataSet**


```python
Order_report=pd.read_excel("Company X - Order Report.xlsx")
courier_invoice=pd.read_excel("Courier Company - Invoice.xlsx")
Company_sku=pd.read_excel("Company X - SKU Master.xlsx")
Company_Pincode=pd.read_excel("Company X - Pincode Zones.xlsx")
rate_chart=pd.read_excel('Courier Company - Rates.xlsx')
```

#**Joining dataset**

```python
# to create a meaningful dataset join data set
Order_invoice=pd.merge(Order_report,courier_invoice, on="Order ID",how="right")
sku=pd.merge(Company_sku,Order_invoice,on='SKU',how='right')
final_df=pd.merge(sku,Company_Pincode,on='Customer Pincode',how='inner' )
final_df
```

```python
# Multiplication of Weight(g)*Order_qty, to find out the actual weight
   #Convertion of gram to Kg of the total weight as per X company

final_df["Total_weight_as_per_X(g)"]=final_df["Order Qty"]*final_df["Weight (g)"]
final_df["Total weight as per X (KG)"]=final_df["Total_weight_as_per_X(g)"]/1000
final_df
```

```python
#for better understanding rename some column name
final_df.rename(columns={'Zone_x':'Delivery Zone charged by Courier Company','Zone_y':'Delivery Zone as per X','Charged Weight':'Total weight as per Courier Company (KG)','Billing Amount (Rs.)':'Charges Billed by Courier Company (Rs.)'}, inplace=True)

final_df
```

```python
# Finding out the weight slab as per company X
condition_1 = [
    (final_df['Total weight as per Courier Company (KG)'] <= 0.500),
    (final_df['Total weight as per Courier Company (KG)'] > 0.500) & (final_df['Total weight as per Courier Company (KG)'] <= 1),
    (final_df['Total weight as per Courier Company (KG)'] > 1) & (final_df['Total weight as per Courier Company (KG)'] <= 1.5),
    (final_df['Total weight as per Courier Company (KG)'] > 1.5) & (final_df['Total weight as per Courier Company (KG)'] <= 2),
    (final_df['Total weight as per Courier Company (KG)'] > 2) & (final_df['Total weight as per Courier Company (KG)'] <= 2.5),
    (final_df['Total weight as per Courier Company (KG)'] > 2.5) & (final_df['Total weight as per Courier Company (KG)'] <= 3),
    (final_df['Total weight as per Courier Company (KG)'] > 3) & (final_df['Total weight as per Courier Company (KG)'] <= 3.5),
    (final_df['Total weight as per Courier Company (KG)'] > 3.5) & (final_df['Total weight as per Courier Company (KG)'] <= 4)
]
value=[0.5,1,1.5,2,2.5,3,3.5,4]
final_df["Weight slab charged by Courier Company (KG)"]=np.select(condition_1,value)

#Finding out the weight slab as per courier company
condition_2 = [
    (final_df['Total weight as per X (KG)'] <= 0.500),
    (final_df['Total weight as per X (KG)'] > 0.500) & (final_df['Total weight as per X (KG)'] <= 1),
    (final_df['Total weight as per X (KG)'] > 1) & (final_df['Total weight as per X (KG)'] <= 1.5),
    (final_df['Total weight as per X (KG)'] > 1.5) & (final_df['Total weight as per X (KG)'] <= 2),
    (final_df['Total weight as per X (KG)'] > 2) & (final_df['Total weight as per X (KG)'] <= 2.5),
    (final_df['Total weight as per X (KG)'] > 2.5) & (final_df['Total weight as per X (KG)'] <= 3),
    (final_df['Total weight as per X (KG)'] > 3) & (final_df['Total weight as per X (KG)'] <= 3.5),
    (final_df['Total weight as per X (KG)'] > 3.5) & (final_df['Total weight as per X (KG)'] <= 4)
]

final_df["Weight slab as per X (KG)"]=np.select(condition_2,value)
#check all the changes done or not
final_df
```

```python
# Function to categorize charge differences
def categorize_charges(charge_diff):
    if charge_diff > 0:
        return 'Overcharged'
    elif charge_diff < 0:
        return 'Undercharged'
    else:
        return 'Correctly Charged'

# Apply the categorize_charges function to create a new column
final_df['Charge Status'] = final_df['Difference Between Expected Charges and Billed Charges (Rs.)'].apply(categorize_charges)

# Print the DataFrame with the new column
final_df
```

```python
charge_counts = final_df.groupby('Charge Status').size()
charge_counts
```

```python
submission_df=pd.DataFrame(charge_counts,columns=['count'])
submission_df.reset_index(inplace=True)
```

```python
submission_df
```

```python
# Calculate total amount for each charge status
charge_status_amounts = final_df.groupby('Charge Status')['Difference Between Expected Charges and Billed Charges (Rs.)'].sum()

# Create a DataFrame for charge status counts and amounts
submission_df = pd.DataFrame({'Count': charge_counts, 'Total Amount': charge_status_amounts})
submission_df
```
